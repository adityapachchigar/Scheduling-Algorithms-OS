```python
from tabulate import tabulate

chart1=[]
chart2=[]
chart3=[]
chart4=[]
class Process:
    def __init__(self,pid,AT,BT):
        self.pid=pid
        self.arrival=AT
        self.burst=BT

def display(chart,burst):
    limit=(5*sum(burst))+1
    for i in range(limit):
        print('-',end="")
    print('')
    for i in range(len(chart)):
        print('| P'+str(chart[i]),end=" ")
    print("|")
    for i in range(limit):
        print('-',end="")
    print("\n\n")

def shiftCL(alist):
    temp=alist[0]
    for i in range(len(alist)-1):
        alist[i]=alist[i+1]
    alist[len(alist)-1]=temp
    return alist

def RR(tq,plist,n):
    global chart1
    queue=[]
    time=0
    ap=0
    rp=0
    done=0
    q=tq
    start=0
    while(done<n):
        for i in range(ap,n):
            if time>=plist[i].arrival:
                queue.append(plist[i])
                ap+=1
                rp+=1
        
        if rp<1:
            chart1.append(0)
            time+=1
            continue

        if start:
            queue=shiftCL(queue)

        if queue[0].burst>0:
            if queue[0].burst>q:
                for g in range(time,time+q):
                    chart1.append(queue[0].pid)
                time+=q
                queue[0].burst-=q
            else:
                for g in range(time,time+queue[0].burst):
                    chart1.append(queue[0].pid)
                time+=queue[0].burst
                queue[0].burst=0
                done+=1
                rp-=1
            start=1

def SRTN(plist,n,preemp):
    queue=[]
    time=0
    ap=0
    rp=0
    done=0

    if not preemp:
        while(done<n):
            for i in range(ap,n):
                if time>=plist[i].arrival:
                    queue.append(plist[i])
                    ap+=1
                    rp+=1
            
            if rp<1:
                time+=1
                chart3.append(0)
                continue

            queue.sort(key=lambda x: (x.burst,x.arrival))

            if queue[0].burst>0:
                chart3.append(queue[0].pid)
                time+=queue[0].burst
                queue[0].burst=99999999
                done+=1
                rp-=1
            
    else:
        while(done<n):
            for i in range(ap,n):
                if time>=plist[i].arrival:
                    queue.append(plist[i])
                    ap+=1
                    rp+=1
            
            if rp<1:
                time+=1
                chart3.append(0)
                continue

            queue.sort(key=lambda x: (x.burst,x.arrival))

            if queue[0].burst>0:
                chart3.append(queue[0].pid)
                time+=1
                queue[0].burst-=1
                if queue[0].burst<1:
                    queue[0].burst=99999999
                    done+=1
                    rp-=1

while True:
    print('\n\n\t\tSCHEDULING ALGORITHMS')
    print('-------------------------------------------------------')
    print('1. FCFS - First Come First Serve - [Non-Preemptive]')
    print('2. RR - Round Robin - [Preemptive]')
    print('3. SRTN - Shortest Remaining Time Next - [Preemptive]')
    print('-------------------------------------------------------')
    ch=int(input('Enter Choice: '))
    if ch==1:
        process=[]
        arrival=[]
        burst=[]
        print('\n')
        n=int(input('Enter No. of Process: '))
        for i in range(n):
            process.append(i+1)
        print('Enter Arrival Time for processes: ')
        for i in range(n):
            ele=int(input())
            arrival.append(ele)
        print('Enter Burst Time for processes: ')
        for i in range(n):
            elem=int(input())
            burst.append(elem)

            
        head=['Finish Time','Turn Around Time','Waiting Time']
        mydata=[]
        FT=[]
        for i in range(n):
            if i==0:
                FT.append(burst[i])
            else:
                FT.append(FT[i-1]+burst[i])
        TAT=[]
        for i in range(n):
            TAT.append(FT[i]-arrival[i])
        WT=[]
        for i in range(n):
            WT.append(TAT[i]-burst[i])
        for i in range(n):
            mydata.append([FT[i],TAT[i],WT[i]])
        # arrival.sort()
        for i in range(n):
            index=arrival.index(min(arrival))
            for j in range(burst[index]):
                chart4.append(process[index])
            arrival[index]+=100
        # print(chart4)
        display(chart4,burst)
        print(tabulate(mydata, headers=head, tablefmt="grid"))
        print('\n')
        print('Average TAT: ',sum(TAT)/n)
        print('Average WT: ',sum(WT)/n)
        print('\n\n')
        
        
    elif ch==2:
        process=[]
        arrival=[]
        burst=[]
        print('\n')
        n=int(input('Enter No. of Process: '))
        for i in range(n):
            process.append(i+1)
        print('Enter Arrival Time for processes: ')
        for i in range(n):
            ele=int(input())
            arrival.append(ele)
        print('Enter Burst Time for processes: ')
        for i in range(n):
            elem=int(input())
            burst.append(elem)
        quantum=int(input('Enter Value of Quantum: '))
        print("\n\n")
        plist=[]
        for i in range(n):
            plist.append(Process(process[i],arrival[i],burst[i]))
        RR(quantum,plist,len(plist))
        display(chart1,burst)
        head=['Finish Time','Turn Around Time','Waiting Time']
        mydata=[]
        FT=[]
        TAT=[]
        WT=[]
        for i in range(n):
            elem=len(chart1)-chart1[::-1].index(process[i])
            FT.append(elem)
        for i in range(n):
            TAT.append(FT[i]-arrival[i])
        for i in range(n):
            WT.append(TAT[i]-burst[i])
        for i in range(n):
            mydata.append([FT[i],TAT[i],WT[i]])
        print(tabulate(mydata, headers=head, tablefmt="grid"))
        print('\n')
        print('Average TAT: ',sum(TAT)/n)
        print('Average WT: ',sum(WT)/n)
        print('\n\n')
            

    elif ch==3:
        process=[]
        arrival=[]
        burst=[]
        print('\n')
        n=int(input('Enter No. of Process: '))
        for i in range(n):
            process.append(i+1)
        print('Enter Arrival Time for processes: ')
        for i in range(n):
            ele=int(input())
            arrival.append(ele)
        print('Enter Burst Time for processes: ')
        for i in range(n):
            elem=int(input())
            burst.append(elem)
        print("\n\n")
        plist=[]
        for i in range(n):
            plist.append(Process(process[i],arrival[i],burst[i]))
        SRTN(plist,len(plist),1)
        display(chart3,burst)
        head=['Finish Time','Turn Around Time','Waiting Time']
        mydata=[]
        FT=[]
        TAT=[]
        WT=[]
        for i in range(n):
            elem=len(chart3)-chart3[::-1].index(process[i])
            FT.append(elem)
        for i in range(n):
            TAT.append(FT[i]-arrival[i])
        for i in range(n):
            WT.append(TAT[i]-burst[i])
        for i in range(n):
            mydata.append([FT[i],TAT[i],WT[i]])
        print(tabulate(mydata, headers=head, tablefmt="grid"))
        print('\n')
        print('Average TAT: ',sum(TAT)/n)
        print('Average WT: ',sum(WT)/n)
        print('\n\n')
    
```
